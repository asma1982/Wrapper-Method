```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = BC.columns.tolist()
for column in cols:
    if BC[column].dtype == 'object':
        BC[column] = le.fit_transform(BC[column])
```

```python
X=BC.drop(["CLASS"],axis=1)
Y=BC["CLASS"]  
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, 
                                                    test_size=0.25)
```

```python
 from sklearn.tree import DecisionTreeClassifier

D_classifier= DecisionTreeClassifier(random_state=0) 


```

```python
rfe = RFE(estimator=D_classifier, step=1)

rfe = rfe.fit(x_train, y_train)
selected_rfe_features = pd.DataFrame({'Feature':list(x_train.columns),
                                      'Ranking':rfe.ranking_})
```

```python
selected_rfe_features.sort_values(by='Ranking')
```

```python
print('Optimal number of features :', rfe.n_features_)
print('Best features :', x_train.columns[rfe.support_])
```

```python
x_train_rfe = rfe.transform(x_train)
x_test_rfe = rfe.transform(x_test)
x_train_rfe[0:3]
```
